```python
import os
from langdetect import detect
```

```python
from datasets import load_dataset

dataset = load_dataset("Open-Orca/OpenOrca", split="train")
```

# Using chain of thoughts instruction tuning data only (english)

```python
dataset[0]
```

```python
cot_only = dataset.filter(lambda example: example["id"].startswith("cot"), num_proc=os.cpu_count())
len(cot_only)
```

```python
cot_non_english = cot_only.filter(lambda example: not detect(example["question"]+example["response"])=="en",num_proc=os.cpu_count())
len(cot_non_english)
```
