## Extract FLAN (800k version)

```python
data = load_dataset("SirNeural/flan_v2", split="train", streaming=True)
```

```python
data_size = 800000 # original_size = 2500000
val_size = 20000
```

```python
extract_data = list(data.take(data_size))
```

```python
# Initialize counters for each task type
cot_count = 0
dialog_count = 0

# Iterate through each item in the list
for item in extract_data:
    # Check if the task key exists and increment the appropriate counter
    if item.get('task') == 'cot':
        cot_count += 1
    elif item.get('task') == 'dialog':
        dialog_count += 1

# Print the results
print(f"Number of items with task 'cot': {cot_count}")
print(f"Number of items with task 'dialog': {dialog_count}")
```

```python
random.seed(42)

# Shuffle the list
random.shuffle(extract_data)

# Calculate split index
split_index = int(0.9 * len(extract_data))  # 90% for training

# Split the list
train_data = extract_data[:split_index]
validation_data = extract_data[split_index:]

# Print the results
print(f"Training data: {len(train_data)}")
print(f"Validation data: {len(validation_data)}")
```

```python
# Initialize counters for each task type
cot_count = 0
dialog_count = 0

# Iterate through each item in the list
for item in train_data:
    # Check if the task key exists and increment the appropriate counter
    if item.get('task') == 'cot':
        cot_count += 1
    elif item.get('task') == 'dialog':
        dialog_count += 1

# Print the results
print(f"Number of items with task 'cot': {cot_count}")
print(f"Number of items with task 'dialog': {dialog_count}")
```

```python
# Initialize counters for each task type
cot_count = 0
dialog_count = 0

# Iterate through each item in the list
for item in validation_data:
    # Check if the task key exists and increment the appropriate counter
    if item.get('task') == 'cot':
        cot_count += 1
    elif item.get('task') == 'dialog':
        dialog_count += 1

# Print the results
print(f"Number of items with task 'cot': {cot_count}")
print(f"Number of items with task 'dialog': {dialog_count}")
```

```python
# Convert to pandas DataFrames
train_df = pd.DataFrame(train_data)
validation_df = pd.DataFrame(validation_data)
```

```python
# Now, convert the DataFrames to Hugging Face's Dataset objects
train_dataset = Dataset.from_pandas(train_df)
validation_dataset = Dataset.from_pandas(validation_df)
```

## Extract FLAN (100k version)

```python
train_data = load_dataset("AlanYky/flan-phi-2-800k", split="train", streaming=True)
```

```python
# Take 100k data (60k Chain of Throught and 40k Dialogs)
cot_count = 0
dialog_count = 0

# Initialize lists to store data points
cot_data = []
dialog_data = []

# Iterate through the dataset
for example in train_data:
    if example['task'] == "cot" and cot_count < 60000:
        cot_data.append(example)
        cot_count += 1
    elif example['task'] == "dialog" and dialog_count < 40000:
        dialog_data.append(example)
        dialog_count += 1
    if cot_count == 60000 and dialog_count == 40000:
        break  # Stop iteration if desired counts are reached

# Concatenate the data points collected for both tasks
all_data = cot_data + dialog_data
```

```python
# Counting the account
cot_count = 0
dialog_count = 0

# Iterate through each item in the list
for item in all_data:
    # Check if the task key exists and increment the appropriate counter
    if item.get('task') == 'cot':
        cot_count += 1
    elif item.get('task') == 'dialog':
        dialog_count += 1

# Print the results
print(f"Number of items with task 'cot': {cot_count}")
print(f"Number of items with task 'dialog': {dialog_count}")
```

```python
# Build a dataframe
df_train = pd.DataFrame(all_data)
```

```python
df_train
```

```python
# Shuffle the dataframe rows
df_train_shuffled = df_train.sample(frac=1).reset_index(drop=True)
```

```python
df_train_shuffled
```

```python
# load the validation (40k)
val_data = load_dataset("AlanYky/flan-phi-2-800k", split="validation", streaming=True)
```

```python
# Take 40k data (20k Chain of Throught and 20k Dialogs)
cot_count = 0
dialog_count = 0

# Initialize lists to store data points
cot_data = []
dialog_data = []

# Iterate through the dataset
for example in val_data:
    if example['task'] == "cot" and cot_count < 20000:
        cot_data.append(example)
        cot_count += 1
    elif example['task'] == "dialog" and dialog_count < 20000:
        dialog_data.append(example)
        dialog_count += 1
    if cot_count == 20000 and dialog_count == 20000:
        break  # Stop iteration if desired counts are reached

# Concatenate the data points collected for both tasks
all_val_data = cot_data + dialog_data


# Initialize counters for each task type
cot_count = 0
dialog_count = 0

# Iterate through each item in the list
for item in all_val_data:
    # Check if the task key exists and increment the appropriate counter
    if item.get('task') == 'cot':
        cot_count += 1
    elif item.get('task') == 'dialog':
        dialog_count += 1

# Print the results
print(f"Number of items with task 'cot': {cot_count}")
print(f"Number of items with task 'dialog': {dialog_count}")
```

```python
# Build a dataframe
df_val = pd.DataFrame(all_val_data)
```

```python
df_val
```

```python
# Shuffle the dataframe rows
df_val_shuffled = df_val.sample(frac=1).reset_index(drop=True)
```

```python
df_val_shuffled
```

```python
# Convert shuffled DataFrames to Hugging Face datasets
train_dataset = Dataset.from_pandas(df_train_shuffled)
val_dataset = Dataset.from_pandas(df_val_shuffled)
```
